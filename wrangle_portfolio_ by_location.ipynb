## Get data

Data from [data.gov](https://catalog.data.gov/dataset/federal-student-loan-portfolio-2e077)

```python
portfolio_df = pd.read_excel('https://data.ed.gov/dataset/fb615e1a-f95f-478d-bf7b-350fdf21133e/resource/e5531678-1689-4cab-a41e-2026c1dc70a9/download/dlportfolio-by-location.xls')
portfolio_df
```

**Change column name**

```python
# Assign the values on index 4 as columns of the data frame
portfolio_df.columns = portfolio_df.iloc[4].values
```

**Remove noizy rows**

```python
# Select only rows from 5 to 45
portfolio_df = portfolio_df.iloc[5:].reset_index(drop=True)
```

```python
portfolio_df.head()
```

```python
# Write the final cleaned data to csv
portfolio_df.to_csv("data/federal_student_aid_portfolio_by_location.csv", mode="w")
```
