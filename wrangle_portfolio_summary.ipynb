## Get data

Data from [data.gov](https://catalog.data.gov/dataset/federal-student-loan-portfolio-2e077)

```python
portfolio_ummary_df = pd.read_excel('https://data.ed.gov/dataset/fb615e1a-f95f-478d-bf7b-350fdf21133e/resource/2596dd40-0d62-4ca1-bce7-ee3ecac13e06/download/portfoliosummary.xls')
portfolio_ummary_df
```

**Change column name**

```python
# Make all the null values No for workability
portfolio_ummary_df = portfolio_ummary_df.fillna("No")
```

```python
# Assign the values on index 4 as columns of the data frame
portfolio_ummary_df.columns = pd.Series(portfolio_ummary_df.iloc[4].values).str.replace("No","School Quarter (Q1-Q4)")
```

**Remove noizy rows**

```python
# Select only rows from 5 to 45
portfolio_ummary_df = portfolio_ummary_df.iloc[5:45]
```
