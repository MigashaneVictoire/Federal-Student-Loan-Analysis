## Get data

Data from [data.gov](https://catalog.data.gov/dataset/federal-student-loan-portfolio-2e077)

```python
portfolio_df = pd.read_excel('https://data.ed.gov/dataset/fb615e1a-f95f-478d-bf7b-350fdf21133e/resource/65239fd0-7d1d-4685-a497-2b4137f59213/download/portfolio-by-debt-size.xls')
portfolio_df
```

**Change column name**

```python
# Make all the null values No for workability
portfolio_df = portfolio_df.fillna("No")
```

```python
# Assign the values on index 4 as columns of the data frame
portfolio_df.columns = pd.Series(portfolio_df.iloc[4].values).str.replace("No","School Quarter (Q1-Q4)")
```

```python
# Engineer a new set of column full column names
cols_vakues = portfolio_df.iloc[3].values
col_index = portfolio_df.iloc[3].index
running_col = "No"
new_col = []

for k,v in enumerate(cols_vakues):
    if v != "No":
        running_col = v
        new_col.append(running_col + " - " + col_index[k])
    if v == "No" and  running_col != "No":
        new_col.append(running_col + " - " + col_index[k])
        running_col = v
    elif v == "No" and  running_col == "No":
        new_col.append(col_index[k])
    
portfolio_df.columns = new_col
```

```python
portfolio_df
```

**Remove noizy rows**

```python
# Select only rows from 5 to 45
portfolio_df = portfolio_df.iloc[5:23].reset_index(drop=True)
```
