## Get data

Data from [data.gov](https://catalog.data.gov/dataset/federal-student-loan-portfolio-2e077)

```python
portfolio_df = pd.read_excel('https://data.ed.gov/dataset/fb615e1a-f95f-478d-bf7b-350fdf21133e/resource/a5ad35a6-2463-4e80-882c-d060c07b1222/download/portfoliobyloantype.xls')
portfolio_df
```

**Change column name**

```python
# Make all the null values No for workability
# portfolio_df = portfolio_df.fillna("No")
```

```python
# Assign the values on index 4 as columns of the data frame
portfolio_df.columns = pd.Series(portfolio_df.iloc[5].values).str.replace("No","School Quarter (Q1-Q4)")
```

**Remove noizy rows**

```python
# Select only rows from 5 to 45
portfolio_df = portfolio_df.iloc[6:35].reset_index(drop=True)
```

```python
portfolio_df
```

**Engineer new columns**

```python
engineered_cols = ['Federal Fiscal Year', 'School Quarter (Q1-Q4)',
       'Stafford Subsidized - Dollars Outstanding (in billions)',
       'Stafford Subsidized -Recipients (in millions)',
       'Stafford Unsubsidized - Dollars Outstanding (in billions)',
       'Stafford Unsubsidized - Recipients (in millions)',
       'Stafford (Combined) - Dollars Outstanding (in billions)',
       'Stafford (Combined) - Loans Recipients (in millions)',
        'Grad PLUS - Dollars Outstanding (in billions)',
       'Grad PLUS - Loans Recipients (in millions)',
       'Parent PLUS - Dollars Outstanding (in billions)',
       'Parent PLUS - Loans Recipients (in millions)',
        'Perkins - Dollars Outstanding (in billions)',
       'Perkins - Loans Recipients (in millions)',
       'Consolidation - Dollars Outstanding (in billions)',
       'Consolidation - Loans Recipients (in millions)']

portfolio_df.columns = engineered_cols
```

```python
# Write the final cleaned data to csv
portfolio_df.to_csv("data/federal_student_aid_portfolio_by_oan_type.csv", mode="w")
```
